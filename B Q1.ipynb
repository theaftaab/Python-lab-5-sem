```python
#import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('iris.csv')
```

```python
data
```

```python
# x_values=data['SepalLength'].values.reshape(-1,1)
x_values = data['SepalLength']
y_values=data['SepalWidth']
```

```python
plt.scatter(x_values,y_values)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot')
plt.show()
```

```python
x_mean=x_values.mean()
y_mean=y_values.mean()
print(f"Mean of X: {x_mean}")
print(f"Mean of Y: {y_mean}")
```

```python
x_values.median()
```

```python
# iii. Calculate median
x_median = np.median(x_values)
y_median = y_values.median()
print(f"Median of X: {x_median}")
print(f"Median of Y: {y_median}")
```

```python
x_std_dev = x_values.std()
y_std_dev = y_values.std()
print(f"Standard Deviation of X: {x_std_dev}")
print(f"Standard Deviation of Y: {y_std_dev}")
```

```python
x_variance = x_values.var()
y_variance = y_values.var()
print(f"Variance of X: {x_variance}")
print(f"Variance of Y: {y_variance}")
```

```python
regression=LinearRegression()

regression.fit(x_values.values.reshape(-1,1),y_values)
slope=regression.coef_[0]
intercept=regression.intercept_
print(f"Slope : {slope}")
```

```python
plt.scatter(x_values,y_values)
plt.plot(x_values,regression.predict(x_values.values.reshape(-1,1)),color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Regression Line')
plt.show()
```
