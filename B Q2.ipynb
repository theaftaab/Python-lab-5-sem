```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the database and displaying the top 10 rows using pandas
file_path = 'tips.csv'  # Replace 'tips.csv' with your file path
data = pd.read_csv(file_path)
```

```python
# Displaying the top 10 rows
print("Top 10 rows of the dataset:")
print(data.head(10))
```

```python
plt.scatter(data['day'],data['tip'])
```

```python
# Scatter Plot (day vs tip)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='day', y='tip', data=data)
plt.title('Scatter Plot of day vs tip')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
# Line Chart (day against tip)
plt.figure(figsize=(8, 6))
sns.lineplot(x='day', y='tip', data=data)
# plt.plot(data['day'],data['tip'])
plt.title('Line Chart of day vs tip')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
# Bar chart with day against tip
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='tip', data=data)
plt.title('Bar Plot of Day against Tip')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
# Histogram of total_bills
plt.figure(figsize=(8, 6))
sns.histplot(data['total_bill'], bins=20, kde=False)
plt.title('Histogram of Total Bills')
plt.xlabel('Total Bills')
plt.ylabel('frequency')
plt.show()
```
